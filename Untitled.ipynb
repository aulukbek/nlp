```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open("vlist.txt")
```

```python
data = []
for line in f:
    data.append(line)
```

```python
df = [line.replace(".", ",").replace("\n", "").split(",")[:2] for line in data]
```

```python
verbs = pd.DataFrame(df, columns= ["infinitive", "past"])
```

```python
import nltk
```

```python
nltk.edit_distance("salam", "salam")
```

```python
import difflib
```

```python
s = difflib.SequenceMatcher(None, "salam", "salam")
```

```python
def similarity(a, b):    
    s = difflib.SequenceMatcher(None, a, b)
    return s.ratio() * 100
```

```python
verbs = verbs.dropna()
sim = verbs[["infinitive", "past"]].apply(lambda i: similarity(i["infinitive"], i["past"]), axis = 1)
```

```python
def diff(a, b):
    s1 = set(a)
    s2 = set(b)
    return s1 ^ s2
```

```python
characters = verbs.apply(lambda i: diff(i["infinitive"], i["past"]), axis = 1)
```

```python
d = pd.DataFrame(characters, columns = ["difference"])
```

```python
edit_dist = verbs.apply(lambda i: nltk.edit_distance(i["infinitive"], i["past"]), axis = 1)
edit_dist.name = "edit_distance"
```

```python
vedit = pd.concat([verbs, edit_dist], axis = 1)
```

```python
vedit[vedit.edit_distance == 0]
```

```python
vedit[vedit.infinitive.str.endswith("e") & vedit.past.str.endswith("d")].to_csv("e_d.csv")
```
