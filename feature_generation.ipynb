```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data_final.csv")
```

```python
df = df[["infinitive", "past", "label", "description"]]
```

```python
# the number of pyhe

features = pd.read_csv("features.csv")
features.columns
```

```python
import pyphen
```

```python
import pyphen
dic = pyphen.Pyphen(lang='en')
```

```python
syllabus = df.infinitive.apply(lambda i: dic.inserted(i)).str.split("-").apply(len)
syllabus.name = "syll_number"
```

```python
last_letter = df.infinitive.str[-1]
last_letter.name = "lst_letter"
```

```python
last_2letter = df.infinitive.str[-2:]
last_2letter.name = "lst_2letter"
```

```python
last_letters = pd.concat([last_letter, last_2letter], axis = 1)
```

```python
# the distributions of vowels 
e_cnt = df.infinitive.str.count("e")
e_cnt.name = "e"

a_cnt = df.infinitive.str.count("a")
a_cnt.name = "a"

i_cnt = df.infinitive.str.count("i")
i_cnt.name = "i"

u_cnt = df.infinitive.str.count("u")
u_cnt.name = "u"

o_cnt = df.infinitive.str.count("o")
o_cnt.name = "o"

oo_cnt = df.infinitive.str.count("oo")
oo_cnt.name = "oo"

ee_cnt = df.infinitive.str.count("ee")
ee_cnt.name = "ee"

vowel_df = pd.concat([e_cnt, a_cnt, i_cnt, u_cnt, o_cnt, oo_cnt, ee_cnt], axis = 1)
vowel_df
```

```python
# the number of vowels 
vowel_cnt = df.infinitive.apply(lambda l: len(set(l).intersection("aeiou")))
vowel_cnt.name = "vowel_cnts"
```

```python
# voiced consonants features 
voiced = ["B", "D", "G", "J", "L", "M", "N", "NG", "R", "SZ", "V", "W", "Y", "Z"]
voiced_ptr = "|".join(voiced)
```

```python
# is the last letter a voiced consonant
#voiced_cons = np.where(df.infinitive.str.upper().str[-1].isin(voiced), 1, 0)
```

```python
voiced_cons = np.where(df.infinitive.str.upper().str[-1].isin(voiced), 1, 0)
voiced_cons
```

```python
# whether last two letters are voiced consonants
voiced_2cons = np.where(df.infinitive.str.upper().str[-2:].str.contains(voiced_ptr), 1, 0)
```

```python
voiced_cnts = pd.concat([pd.Series(voiced_cons), pd.Series(voiced_2cons)], axis = 1)
voiced_cnts.columns = ["voiced_lst", "voiced_2lst"]
voiced_cnts
```

```python
voiceless = ["CH", "F", "K", "P", "S", "SH", "T" ]
voiceless_ptr = "|".join(voiceless)
voiceless_ptr
```

```python
voiceless_2cons = np.where(df.infinitive.str.upper().str[-2:].str.contains(voiceless_ptr), 1, 0)
voiceless_cons = np.where(df.infinitive.str.upper().str[-1].isin(voiceless), 1, 0)
voiceless_df = pd.concat([pd.Series(voiceless_cons), pd.Series(voiceless_2cons)], axis = 1)
voiceless_df.columns = ["voiceless_lst", "voiceless_2lst"]
voiceless_df
```

```python
consonants = pd.concat([voiced_cnts, voiceless_df] )
```

```python
consonants = voiceless_df.join(voiced_cnts)
```

```python
# d t c 

d_ = df.infinitive.str.endswith("d").astype(int)
t_ = df.infinitive.str.endswith("t").astype(int)
c_ = df.infinitive.str.endswith("c").astype(int)

dtc = pd.concat([d_, t_, c_], axis = 1)
dtc.columns = ["d_", "t_", "c_"]
```

```python
features = pd.concat([vowel_df, consonants, dtc, vowel_cnt, last_letters, syllabus, df.label], axis = 1)
```
